# Цель данного отчета 
Рассмотреть функции активации, которые существуют в библиотеке Keras. Будут представлены их названия, математические и графические представления, а также особенности использования, достоинства и недостатки. Также будет проведена визуализация функций активации на одном графике.

## ELU (Exponential Linear Unit):
Математическое представление: Функция активации ELU представляет собой комбинацию линейной и экспоненциальной функций. Для положительных значений входа x, ELU(x) = x, а для отрицательных значений, ELU(x) = alpha * (exp(x) - 1), где alpha - параметр, определяющий отрицательную область функции.

Графическое представление: График функции ELU имеет линейный участок для положительных значений и экспоненциальный участок для отрицательных значений.

Особенности использования и достоинства: ELU является альтернативой функции активации ReLU. Она обеспечивает отсутствие проблемы "мертвых нейронов" и имеет отрицательные значения для отрицательных входов, что помогает бороться с проблемой исчезающего градиента. ELU также способствует более быстрой сходимости при обучении нейронных сетей.

Программная реализация: keras.activations.elu()

## Softmax:

Математическое представление: Функция активации Softmax преобразует вектор входных значений в вероятностное распределение. Она вычисляет экспоненциальную функцию от каждого элемента вектора и нормализует результаты, чтобы их сумма была равна 1.

Графическое представление: График функции Softmax представляет собой S-образную кривую, где значения стремятся к 0 или 1.

Особенности использования и достоинства: Softmax обычно применяется в последнем слое многоклассовой классификации, где требуется предсказать вероятности принадлежности к различным классам. Она гарантирует, что сумма вероятностей равна 1, что позволяет интерпретировать вывод модели как распределение вероятностей.

Программная реализация: keras.activations.softmax()

## SELU (Scaled Exponential Linear Unit):

Математическое представление: Функция активации SELU является модификацией ELU и определена как SELU(x) = scale * (alpha * exp(x) - alpha), если x < 0, и SELU(x) = scale * x, если x >= 0, где alpha и scale - параметры, выбираемые для обеспечения нормализации.

Графическое представление: График функции SELU похож на ELU, но с масштабированием и нормализацией значений.
Особенности использования и достоинства: SELU обладает свойством самонормализации, что позволяет сохранять статистическую норму входных данных при применении глубоких нейронных сетей. Она может улучшить качество обучения и предотвратить взрывной или затухающий градиент.

Программная реализация: keras.activations.selu()

## Softplus:

Математическое представление: Функция активации Softplus определяется как Softplus(x) = log(1 + exp(x)).

Графическое представление: График функции Softplus имеет форму плавно нарастающей кривой.

Особенности использования и достоинства: Softplus является гладкой аппроксимацией функции активации ReLU. Она может быть полезной в задачах регрессии, где требуется ограничить значения выхода модели сверху.

Программная реализация: keras.activations.softplus()

## Softsign:

Математическое представление: Функция активации Softsign определена как Softsign(x) = x / (1 + |x|).

Графическое представление: График функции Softsign представляет собой S-образную кривую.

Особенности использования и достоинства: Softsign имеет свойство гладкости и непрерывности. Она может быть полезной в нейронных сетях с большими значениями входных данных, так как она ограничивает значения выхода в диапазоне (-1, 1).

Программная реализация: keras.activations.softsign()

## ReLU (Rectified Linear Unit):

Математическое представление: Функция активации ReLU определена как ReLU(x) = max(0, x).

Графическое представление: График функции ReLU имеет линейный участок для положительных значений и нулевое значение для отрицательных значений.

Особенности использования и достоинства: ReLU является одной из самых популярных функций активации. Она линейно активирует положительные значения и обнуляет отрицательные значения. ReLU проста в реализации и обладает высокой эффективностью вычислений, но может привести к проблеме "мертвых нейронов" при которой некоторые нейроны могут остановиться в активации 0.

Программная реализация: keras.activations.relu()

## Tanh (Hyperbolic Tangent):

Математическое представление: Функция активации Tanh определена как Tanh(x) = (exp(x) - exp(-x)) / (exp(x) + exp(-x)).

Графическое представление: График функции Tanh представляет собой S-образную кривую с диапазоном значений от -1 до 1.

Особенности использования и достоинства: Tanh является симметричной функцией с нулевым центром, что позволяет модели обрабатывать отрицательные значения входных данных. Она может быть полезна в задачах, где требуется предсказание значений в диапазоне (-1, 1).

Программная реализация: keras.activations.tanh()

## Sigmoid:

Математическое представление: Функция активации Sigmoid определена как Sigmoid(x) = 1 / (1 + exp(-x)).

Графическое представление: График функции Sigmoid представляет собой S-образную кривую, где значения стремятся к 0 или 1.

Особенности использования и достоинства: Sigmoid применяется в задачах бинарной классификации, где требуется предсказать вероятность принадлежности к одному из двух классов. Она ограничивает значения выхода в диапазоне (0, 1) и может использоваться для моделирования нелинейных зависимостей в данных.

Программная реализация: keras.activations.sigmoid()

## Hard Sigmoid:

Математическое представление: Функция активации Hard Sigmoid определена как HardSigmoid(x) = max(0, min(1, (x * 0.2) + 0.5)).

Графическое представление: График функции Hard Sigmoid представляет собой линейную функцию с насыщением в точках 0 и 1.

Особенности использования и достоинства: Hard Sigmoid является быстрой аппроксимацией Sigmoid с целью ускорить вычисления в нейронных сетях. Она может быть полезна в задачах, где скорость вычислений критична.

Программная реализация: keras.activations.hard_sigmoid()

## Exponential:

Математическое представление: Функция активации Exponential определена как Exponential(x) = exp(x).

Графическое представление: График функции Exponential является восходящей экспоненциальной кривой.

Особенности использования и достоинства: Exponential преобразует входные значения в экспоненциальную форму, что может быть полезным в задачах, где требуется усилить различия между значениями и подчеркнуть большие значения.

Программная реализация: keras.activations.exponential()

## Linear:

Математическое представление: Функция активации Linear определена как Linear(x) = x.

Графическое представление: График функции Linear представляет собой прямую линию с углом наклона 45 градусов.

Особенности использования и достоинства: Linear не вносит нелинейных преобразований и является самой простой функцией активации. Она может быть полезна в задачах, где требуется линейное преобразование входных данных.

Программная реализация: keras.activations.linear()

## Step:

Математическое представление: Функция активации Step определена как Step(x) = 1, если x >= 0, и Step(x) = 0, если x < 0.

Графическое представление: График функции Step представляет собой горизонтальную линию, которая переходит из 0 в 1 при x = 0.

Особенности использования и достоинства: Step применяется в задачах бинарной классификации, где требуется принимать решение на основе порогового значения. Она может быть полезна при простых задачах, где не требуется учитывать нюансы межклассовых отношений.

Программная реализация: keras.activations.step()

# Визуализация функций активации на одном графике:

```python
import numpy as np
import matplotlib.pyplot as plt

# Задание диапазона значений x
x = np.linspace(-5, 1, 100)

# Функции активации
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / np.sum(e_x)

def linear(x):
    return x

def elu(x):
    return np.where(x > 0, x, np.exp(x) - 1)

def softplus(x):
    return np.log(1 + np.exp(x))

def softsign(x):
    return x / (1 + np.abs(x))

def hard_sigmoid(x):
    return np.clip(0.2 * x + 0.5, 0, 1)

def exponential(x):
    return np.exp(x)

def step(x):
    return np.where(x >= 0, 1, 0)

def selu(x):
    return np.where(x > 0, x, np.exp(x) - 1)

# Вычисление значений функций активации
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_softmax = softmax(x)
y_linear = linear(x)
y_elu = elu(x)
y_softplus = softplus(x)
y_softsign = softsign(x)
y_hard_sigmoid = hard_sigmoid(x)
y_exponential = exponential(x)
y_step = step(x)
y_selu = selu(x)

# Построение графика
plt.figure(figsize=(10, 6))

plt.plot(x, y_sigmoid, label='Sigmoid')
plt.plot(x, y_tanh, label='Tanh')
plt.plot(x, y_relu, label='ReLU')
plt.plot(x, y_softmax, label='Softmax')
plt.plot(x, y_linear, label='Linear')
plt.plot(x, y_elu, label='ELU')
plt.plot(x, y_softplus, label='Softplus')
plt.plot(x, y_softsign, label='Softsign')
plt.plot(x, y_hard_sigmoid, label='Hard Sigmoid')
plt.plot(x, y_exponential, label='Exponential')
plt.plot(x, y_step, label='Step', c='limegreen')
plt.plot(x, y_selu, label='SELU', c='lightblue')

# Настройка легенды и подписей осей
plt.legend()
plt.xlabel('x')
plt.ylabel('Activation')

# Отображение графика
plt.show()
```

# Собственная нейросеть

```python
import numpy as np

# Входные данные (ширина и длина)
X = np.array([[8, 10], [5, 20]])

# Метки классов (0 - коровка, 1 - гусеница)
y = np.array([0, 1])

# Функция потерь (бинарная кросс-энтропия)
def binary_crossentropy(y_true, y_pred):
    return -(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# Инициализация весов
weights = np.random.randn(2)

# Прямое распространение
def forward(X, weights):
    return 1 / (1 + np.exp(-(np.dot(X, weights))))

# Обучение модели
epochs = 500
learning_rate = 0.01

for epoch in range(epochs+1):
    # Прямое распространение
    y_pred = forward(X, weights)
    
    # Вычисление функции потерь
    loss = binary_crossentropy(y, y_pred)
    
    # Вычисление градиента
    gradient = np.dot(X.T, (y_pred - y))
    
    # Обновление весов с использованием градиентного спуска
    weights -= learning_rate * gradient
    
    # Вывод значения функции потерь на каждой эпохе
    if epoch % 25 == 0:
        print(f"Эпоха: {epoch+1} --- Потеря: {loss}")

# Тестирование модели
X_test = np.array([[8, 10], [5, 20], [6, 15], [7, 15]])  # Новые данные для тестирования
y_pred_test = np.array([0, 1, 1, 0])
predictions = [0 if pred < 0.5 else 1 for pred in y_pred_test]

# Вывод результатов тестирования
print('\n')
for i in range(len(X_test)):
    print(f"{i+1}) Входные данные: {X_test[i]}, Предсказанный класс: {predictions[i]}")
```
