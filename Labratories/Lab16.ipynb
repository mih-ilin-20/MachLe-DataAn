```python
p = 67
points = []

for x in range(p):
    y_squared = (x**3 + x + 1) % p
    # Проверяем, является ли y_squared квадратом по модулю p
    for y in range(p):
        if (y**2) % p == y_squared:
            points.append((x, y))

print(f"Всего точек: {len(points)}")
for i, point in enulambderate(points, start=1):
    print(f"K = {i}  {point}")
```

* В цикле for x in range(p) перебираем все возможные значения x.
* x_squared - вычисляем значение y^2 по уравнению эллиптической кривой.
* Проверяем, является ли x_squared квадратичным вычетом.
* Если является, вычисляем y с использованием выражения для корня квадратного.
* Добавляем точку (x, y) и её отражение (x, p - y) относительно оси x в список точек.
* Возвращаем найденные точки.

В конце кода выводится список всех точек на эллиптической кривой. Это список, который мы будем использовать для выполнения следующих заданий.

```python
p = 67
a = 1
b = 1

def point_addition(p1, p2, a, p):
    # Сложение двух точек на эллиптической кривой
    if p1 == (float('inf'), float('inf')):
        return p2
    if p2 == (float('inf'), float('inf')):
        return p1

    x1, y1 = p1
    x2, y2 = p2

    if p1 != p2:
        # Вычисление наклона (slope) для несовпадающих точек
        m = ((y2 - y1) * pow(x2 - x1, -1, p)) % p
    else:
        if y1 == 0:
          return (float('inf'), float('inf'))  # точка на бесконечности
        # Вычисление наклона для совпадающих точек
        m = ((3 * x1**2 + a) * pow(2 * y1, -1, p)) % p

    # Вычисление x и y для новой точки
    x3 = (m**2 - x1 - x2) % p
    y3 = (m * (x1 - x3) - y1) % p

    return (x3, y3)


def point_mul(point, scalar, a, p):
    result = (float('inf'), float('inf'))
    for _ in range(scalar.bit_length()):
        if scalar & 1:
            result = point_addition(result, point, a, p)
        point = point_addition(point, point, a, p)
        scalar >>= 1
    return result

# Заданные точки
point1 = (58, 0)
point2 = (7, 4)
point5 = (34, 12)


# Вычисление суммы точек
sum_point1_2 = point_addition(point1, point2, a, p)
sum_point2_3 = point_addition(point2, point2, a, p)
point_5 = point_mul(point5, 28, a, p)

# Вывод результатов
print("Сумма точек {58, 0} и {7, 4}:", sum_point1_2)
print("Сумма точек {7, 4} и {7, 4}:", sum_point2_3)
print("Точка 28 {34, 12}:", point_5)
```
